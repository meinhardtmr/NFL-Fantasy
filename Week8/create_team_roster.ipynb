### Add QB and RB

```python
list = []

for q in qb.index:
#    week = qb['WEEK'][q]
    qb_name = qb['NICKNAME'][q]
#    qb_team = qb['TEAM'][q]
#    qb_game = qb['GAME'][q]
    qb_salary = qb['SALARY'][q]
    qb_matchup = qb['MATCHUP'][q]
    qb_fpts = qb['PROJ_FPTS'][q]
    for r in rb.index[0:]:
        rb1 = rb['NICKNAME'][r]
#        rb1_team = rb['TEAM'][r]
#        rb1_game = rb['GAME'][r]
        rb1_salary = rb['SALARY'][r]
        rb1_matchup = rb['MATCHUP'][r]
        rb1_fpts = rb['PROJ_FPTS'][r]
        for r2 in rb.index[r+1:]:
            rb2 = rb['NICKNAME'][r2]
#            rb2_team = rb['TEAM'][r2]
#            rb2_game = rb['GAME'][r2]
            rb2_salary = rb['SALARY'][r2]
            rb2_matchup = rb['MATCHUP'][r]
            rb2_fpts = rb['PROJ_FPTS'][r2]
            list.append([#week,
                         qb_name,
                         #qb_salary,
                         #round(qb_fpts,2), 
                         rb1,
                         #rb1_salary,
                         #round(rb1_fpts,2), 
                         rb2,
                         #rb2_salary,
                         #round(rb2_fpts), 
                         (qb_salary + rb1_salary + rb2_salary),
                         round(qb_fpts + rb1_fpts + rb2_fpts,2),
                         qb_matchup + rb1_matchup + rb2_matchup])

df_1 = pd.DataFrame(list, columns=[#'week',
                                   'qb',
                                   #'qb_salary',
                                   #'qb_fpts',
                                   'rb1',
                                   #'rb1_salary',
                                   #'rb1_fpts',
                                   'rb2',
                                   #'rb2_salary',
                                   #'rb2_fpts',
                                   'total_salary',
                                   'total_fpts',
                                   'total_matchup'])

print(len(df_1.index))
df_1.head()
```

```python
df_1.to_excel(Path('week6_qb_rb.xlsx'), index=False)
```

### Aggregate WR

```python
list = []

for w in wr.index:
    wr1 = wr['NICKNAME'][w]
#    wr1_team = wr['TEAM'][w]
#    wr1_game = wr['GAME'][w]
    wr1_salary = wr['SALARY'][w]
    wr1_matchup = wr['MATCHUP'][w]
    wr1_fpts = wr['PROJ_FPTS'][w]
    for w2 in wr.index[w+1:len(wr.index)]:
        wr2 = wr['NICKNAME'][w2]
#        wr2_team = wr['TEAM'][w2]
#        wr2_game = wr['GAME'][w2]
        wr2_salary = wr['SALARY'][w2]
        wr2_matchup = wr['MATCHUP'][w2]
        wr2_fpts = wr['PROJ_FPTS'][w2]
        for w3 in wr.index[w2+1:len(wr.index)]:
            wr3 = wr['NICKNAME'][w3]
#            wr3_team = wr['TEAM'][w3]
#            wr3_game = wr['GAME'][w3]
            wr3_salary = wr['SALARY'][w3]
            wr3_matchup = wr['MATCHUP'][w3]
            wr3_fpts = wr['PROJ_FPTS'][w3]
            list.append([wr1,
#                         wr1_salary,
#                         round(wr1_fpts,2),
                         wr2,
#                         wr2_salary,
#                         round(wr2_fpts,2),
                         wr3,
#                         wr3_salary,
#                         round(wr3_fpts,2),
                         wr1_salary + wr2_salary + wr3_salary,
                         round(wr1_fpts + wr2_fpts + wr3_fpts,2),
                         wr1_matchup + wr2_matchup + wr3_matchup])

df_wr = pd.DataFrame(list, columns=['wr1',
#                                    'wr1_salary',
#                                    'wr1_fpts',
                                    'wr2',
#                                    'wr2_salary',
#                                    'wr2_fpts',
                                    'wr3',
#                                    'wr3_salary',
#                                    'wr3_fpts',
                                    'total_salary',
                                    'total_fpts',
                                    'total_matchup'])  

print(len(df_wr.index))
df_wr.head()
```

### Combine QB, RB and WR

```python
cnt_1 = 0
cnt_2 = 0
print(dt.today())

list = []
for r in df_1.index:
    for w in df_wr.index:
        cnt_1 += 1
        if df_1['total_salary'][r] + df_wr['total_salary'][w] <= 60000:
            cnt_2 += 1
            list.append([#df_1['week'][r],
                         df_1['qb'][r],
                         #df_1['qb_salary'][r],
                         #df_1['qb_fpts'][r],
                         df_1['rb1'][r],
                         #df_1['rb1_salary'][r],
                         #df_1['rb1_fpts'][r],
                         df_1['rb2'][r],
                         #df_1['rb2_salary'][r],
                         #df_1['rb2_fpts'][r],
                         df_wr['wr1'][w],
                         #df_wr['wr1_salary'][w],
                         #df_wr['wr1_fpts'][w],
                         df_wr['wr2'][w],
                         #df_wr['wr2_salary'][w],
                         #df_wr['wr2_fpts'][w],
                         df_wr['wr3'][w],
                         #df_wr['wr3_salary'][w],
                         #df_wr['wr3_fpts'][w],
                         df_1['total_salary'][r] + df_wr['total_salary'][w],
                         round(df_1['total_fpts'][r] + df_wr['total_fpts'][w],2),
                         df_1['total_matchup'][r] + df_wr['total_matchup'][w]])
            
df_rb_wr = pd.DataFrame(list, columns=[#'week',
                                       'qb',
                                       #'qb_salary',
                                       #'qb_fpts',
                                       'rb1',
                                       #'rb1_salary',
                                       #'rb1_fpts',
                                       'rb2',
                                       #'rb2_salary',
                                       #'rb2_fpts',
                                       'wr1',
                                       #'wr1_salary',
                                       #'wr1_fpts',
                                       'wr2',
                                       #'wr2_salary',
                                       #'wr2_fpts',
                                       'wr3',
                                       #'wr3_salary',
                                       #'wr3_fpts',
                                       'total_salary',
                                       'total_fpts',
                                       'total_matchup'])       

print(dt.today())
print(len(df_rb_wr.index), cnt_1, cnt_2)
df_rb_wr.head()
# 216600 216600 216600
#216600 216600 216600
```

### Merge Files

```python
df_fan = pd.read_csv('FanDuel-NFL-2023 ET-10 ET-15 ET-94704-players-list.csv')
df_dst = pd.read_csv('FantasyPros_2023_Week_6_DST_Rankings.csv')
df_qb = pd.read_csv('FantasyPros_2023_Week_6_QB_Rankings.csv')
df_rb = pd.read_csv('FantasyPros_2023_Week_6_RB_Rankings.csv')
df_te = pd.read_csv('FantasyPros_2023_Week_6_TE_Rankings.csv')
df_wr = pd.read_csv('FantasyPros_2023_Week_6_WR_Rankings.csv')
```

```python
df_concat = pd.concat([df_qb,df_rb, df_wr, df_te, df_dst])
df_concat.rename(columns={'PLAYER NAME':'Nickname'}, inplace=True)
df_concat.head()
```

```python
df_merge = pd.merge(df_fan, df_concat, how='left')
df_merge.head()
```
